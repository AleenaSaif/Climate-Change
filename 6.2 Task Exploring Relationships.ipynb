### Exploratory Visual Analysis

#### Creating a Subset for Visual Analysis

```python
# Create a subset df

df_sub = df[['year', 'temp_change' ]]
```

```python
# Check subset

df_sub.columns
```

### Correlation Matrix Heatmap

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

Temperature change has a weak correlation with year.

```python
corr.figure.savefig(os.path.join(path, 'Visualizations', 'corr_heatmap.png' ))
```

### Scatter Plot

```python
# Create a scatterplot for the 'year' and 'temp_change' columns in seaborn

b = sns.lmplot(x = 'year', y = 'temp_change', data = df)
```

```python
b.figure.savefig(os.path.join(path, 'Visualizations', 'scatter_plot.png' ))
```

### Pair Plots

```python
# Create a pair plot 

a = sns.pairplot(df_sub)
```

```python
a.figure.savefig(os.path.join(path, 'Visualizations', 'Pair_plot.png' ))
```

### Categorical Plot

#### Creating Seasons Category

```python
df.loc[df['months'] == 'Spring', 'Season'] = 'Spring'
```

```python
df.loc[df['months'] == 'Winter', 'Season'] = 'Winter'
```

```python
df.loc[df['months'] == 'Fall', 'Season'] = 'Fall'
```

```python
df.loc[df['months'] == 'Summer', 'Season'] = 'Summer'
```

```python
# Checking column 'Season'

df['Season'].value_counts(dropna = False)
```

```python
# Check dataset

df.columns
```

```python
# Create Categorical Plot

sns.set(style="ticks")
g = sns.catplot(x="Season", y="temp_change", hue="year", data=df)
```

```python
g.figure.savefig(os.path.join(path, 'Visualizations', 'categorical_plot.png' ))
```

### Exploratory Answers

#### Is there any significant difference between seasons

#### Categorical Plot shows the difference in temperature change during seasons. Summer & Fall season has a relatively stable temperature change. Spring has a relatively rapid change in temperature. Whereas, Winter has the most volatile change among all the seasons.
